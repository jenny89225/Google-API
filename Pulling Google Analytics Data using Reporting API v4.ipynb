```python
import pandas as pd
from google2pandas import *

## write query to fetch ga data 
query1 = {
    'reportRequests': [{
        'viewId' : 'xxxxx', ## google analytics view id
        
        'dateRanges': [{
            'startDate' : '2021-01-01',
            'endDate'   : '2021-06-26'}],

        'metrics'   : [
            {'expression': 'ga:pageviews'},
            {'expression': 'ga:totalEvents'},
            {'expression': 'ga:sessions'},
            {'expression': 'ga:users'},
            {'expression': 'ga:newUsers'}, 
             ],
            
        'dimensions': [
              {'name': 'ga:sourceMedium'},
              {'name': 'ga:pagePath'},
              {'name': 'ga:country'},
              {'name': 'ga:operatingSystem'},
              {'name': 'ga:mobileDeviceModel'},
#               {'name': 'ga:eventCategory'},
#               {'name': 'ga:eventAction'},
#               {'name': 'ga:eventLabel'},
                      ],
        'orderBys': [
            {"fieldName": "ga:sessions", "sortOrder": "DESCENDING"},
            {"fieldName": "ga:pageviews", "sortOrder": "DESCENDING"}
       ],
#         "dimensionFilterClauses": [
#         {
#           "filters": [
#             {
#               "dimensionName": "ga:dimension7",
#               "operator": "PARTIAL",
#               "expressions": ["66910122225119"]
#             }
#           ]
#         }
#       ],
  }]  
}
    
conn = GoogleAnalyticsQueryV4(secrets='xxxxxx')  ## Service Account credentials

query1_df = conn.execute_query(query1)

query1_df.tail(10)
```
